```python
import speech_recognition as sr

# Initialize recognizer
recognizer = sr.Recognizer()

# Function to recognize speech from an uploaded audio file
def audio_to_text(audio_file_path):
    # Open the audio file
    with sr.AudioFile(audio_file_path) as source:
        # Adjust for ambient noise and record the audio
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.record(source)

    try:
        # Use Google's Web Speech API to recognize the speech in the audio
        text = recognizer.recognize_google(audio)
        print("Text from audio: ", text)
        return text
    except sr.UnknownValueError:
        print("Sorry, the speech was not recognized.")
        return None
    except sr.RequestError:
        print("Could not request results from Google Speech Recognition service.")
        return None

# Replace 'path_to_audio_file.wav' with the uploaded file name
audio_file_path = "03-01-03-01-01-01-01.wav"  # Use your file name here
audio_to_text(audio_file_path)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load the model and tokenizer
model_name = "Salesforce/codegen-350M-mono"  # lightweight for demo
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

def generate_code(prompt):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(
        **inputs,
        max_length=256,
        do_sample=True,
        temperature=0.7,
        top_p=0.95,
        pad_token_id=tokenizer.eos_token_id
    )
    code = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return code.replace(prompt, "").strip()

# Ask for user input
user_prompt = input("Enter your coding prompt: ")
generated_code = generate_code(user_prompt)

print("\n🧾 Generated Code:\n")
print(generated_code)
```

4. entity name extraction

```python
import spacy

# Load the pre-trained spaCy model
nlp = spacy.load("en_core_web_sm")

def extract_entities(text):
    # Process the text
    doc = nlp(text)

    # Extract entities
    entities = [(entity.text, entity.label_) for entity in doc.ents]

    return entities

# Example usage
sentence = "Elon Musk is the CEO of Tesla, and he lives in California."
entities = extract_entities(sentence)

print(entities)
```
